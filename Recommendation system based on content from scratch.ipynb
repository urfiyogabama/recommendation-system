```python
import pandas as pd
import numpy as np
from numpy.random import *
import nltk as nl
from nltk.corpus import stopwords
from nltk.stem import PorterStemmer 
import math

ps = PorterStemmer()

data=pd.read_csv("udemy_courses2.csv")
```

```python
X=data.values

n=len(data)
z=np.zeros((n))
for i in range (n):
    if X[i,7]=='Business Finance':
        z[i]=1
    elif X[i,7]=='Graphic Design':
        z[i]=2
    elif X[i,7]=='Musical Instruments':
        z[i]=3
    else:
        z[i]=4
np.savetxt('myfile.csv', z, delimiter=',')

L=np.zeros((n))
for i in range (n):
    if X[i,5]=='Beginner Level':
        L[i]=0
    elif X[i,5]=='All Levels':
        L[i]=1
    elif X[i,5]=='Intermediate Level':
        L[i]=2
    else:
        L[i]=3
```

```python
data['Answer Key']=z
data['No. Level']=L
```

```python
X=data.values
corpus=[]
for i in range (len(X)):
    doc=nl.word_tokenize(X[i,1])
    corpus.append([doc])
```

```python
# cleaning the data
# 1 Menghapus tanda2 non alfabet

for j in range(len(corpus)):
    idx=0
    for i in corpus[j]:
        i=[word for word in i if word.isalpha()]
        corpus[j][idx]=i
        idx+=1
    # 2 delete stopwords
    idx=0
    for i in corpus[j]:
        i=[w for w in i if not w in stopwords.words('english')]
        corpus[j][idx]=i
        idx+=1
    # 3 stem the data -> jadi kata dasar
    idx=0
    for i in corpus[j]:
        i=[ps.stem(w) for w in i]
        corpus[0][idx]=i
        idx+=1
```

```python
judul=[]
for i in range (len(corpus)):
    jud=' '.join(corpus[i][0])
    judul.append(jud)
#corpus=[]
#for i in range (len(judul)):
 #   corpus.append([judul[i][0]])
```

```python
# Mengumpulkan kata apa saja yang ada pada keseluruhan judul
# Create an empty dictionary 
d = dict()  
for line in judul: 
    # Split the line into words 
    words = line.split(" ") 
    # Iterate over each word in line
    for word in words: 
        # Check if the word is already in dictionary 
        if word in d: 
        # Increment count of word by 1 
           d[word] = d[word] + 1
        else: 
        # Add the word to dictionary with count 1 
           d[word] = 1
for key in list(d.keys()): 
    print(key, ":", d[key])
```

```python
d_all=np.zeros((len(d),len(judul)))
d_n=0
def TF(corpus):
    y=0
    global d
    d1 = dict.fromkeys(list(d), y)
    v1=[corpus]
    for line in v1: 
        # Split the line into words 
        words = line.split(" ") 
        # Iterate over each word in line 
        for word in words: 
            # Check if the word is already in dictionary 
            if word in d1: 
            # Increment count of word by 1 
               d1[word] = d1[word] + 1
            #else: 
            # Add the word to dictionary with count 1 
             #  d1[word] = 1
        n=len(words)
    global d_all
    global d_n
    val=d1.values()
    a=list(val)
    d_all[:,d_n]=list(val) #untuk idf
    d_n+=1
    tf=np.zeros((len(d1),1))
    for i in range (len(a)): 
        #print(key, ":", d1[key])
        tf[i]= a[i]/n
    return tf,d_all,d1
```

```python
#Calling the funtion
tfall=[]
for i in range (len(judul)):
    for ver in [judul[i]]:
        tf,d_all,d1=TF(ver)
        tfall.append([tf])
    
d_vektor=np.zeros((len(d),1))
for i in range (len(d)):
    for j in range (len(judul)):
        if d_all[i,j]!=0:
            d_vektor[i]+=1
idf=np.zeros((len(d),1))
tfidf=[]
tfidfall=[]
for i in range (len(d)):
    idf[i]=math.log(len(judul)/d_vektor[i]) 
for i in range (len(judul)):
    tfidf=tfall[i]*idf
    tfidfall.append([tfidf])
def vektor(x):
    a=x[0]
    b=a[0]
    return b
```

```python
#vektor TFID
vek=[]
for i in range (len (tfidfall)):
    v=vektor(tfidfall[i])
    vek.append([v])
```
